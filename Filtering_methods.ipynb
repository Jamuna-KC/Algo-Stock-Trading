# Applying filters to identify reliable stocks

Importing libraries and dataset

```python
import pandas as pd
df=pd.read_csv("/content/stock_data.csv")
df.iloc[:,:]
df
```

selecting stocks which lies under certian current market price(CPM) 500 in this case


> The current market price (CMP) is the most recent price at which an asset or security was traded in the open market.



```python
df=df[df.CMP_Rs<500]
df
```

Filtering according to NPM previous year and NPM last quarter


> Net Profit Margin (NPM) for the previous year is the ratio of net profit to total revenue for the entire preceding fiscal year. NPM for the last quarter is the same ratio, but calculated for the most recent three-month period.



```python
df=df[df.NPM_Ann_Pr>=10]
df=df[df.NPM_Qtr_Pr>=10]
df
```

Filtering stocks with Return on invested capital more than certain value


> Return on Invested Capital (ROIC) is a financial metric that measures the profitability of a company by evaluating its net operating profit in relation to the total capital employed.



```python
df=df[df.ROIC_Pr>=10]
df
```

Filtering stocks based on Average return on equity 5yrs


> Average Returns on Equity over five years represent the mean percentage increase or decrease in a company's shareholder equity annually, providing insights into its long-term financial performance.



```python
df=df[df.ROE_5Yr_Pr>=15]
df
```

Filtering according to ROCE


> Return on Capital Employed (ROCE) is a financial metric that evaluates the efficiency of a company in generating profits from its capital, calculated by dividing operating profit by the capital employed.



```python
df=df[df.ROCE_Pr >=15]
df
```

Filtering according to EPS

> Earnings Per Share (EPS) is a financial metric representing the portion of a company's profit allocated to each outstanding share of common stock, providing insight into its profitability on a per-share basis.



```python
df=df[df.EPS_12M_Rs	>=10]
df
```

Filtering accoring to average volume purchase based on last month

> Average Volume Purchase in stocks is the mean quantity of shares traded over a specified period, reflecting the average level of market activity for buying transactions during that timeframe.



```python
df=df[df.Avg_Vol_1Mth>=1_00_000]
df
```

Therefore, through the application of the aforementioned filters, we can pinpoint robust stocks that serve as a secure foundation for constructing investment algorithms. This method hence helps us to identify stocks which are profitable for future investments.
